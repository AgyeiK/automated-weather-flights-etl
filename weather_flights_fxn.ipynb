```python
import requests
import pandas as pd

def get_weather_dataframe(cities):
    """
    Fetches location and current weather for multiple cities and returns a DataFrame.

    Parameters:
    - cities: list of city names (strings)
    - country: country code (default "DE")
    - API_key: your OpenWeather API key

    Returns:
    - pd.DataFrame with columns: name, country, lat, lon, temp, feels_like, weather, humidity, wind_speed
    """
    data = []

    for city in cities:
        # 1️Geocoding: get lat/lon
        geo_url = "https://api.openweathermap.org/geo/1.0/direct"
        API_key_geo=os.getenv("API")
        geo_params = {"q": f"{city},{country}", "limit": 1, "appid": API_key_geo}
        geo_response = requests.get(geo_url, params=geo_params)
        geo_json = geo_response.json()

        if isinstance(geo_json, list) and len(geo_json) > 0:
            city_data = geo_json[0]
            lat = city_data["lat"]
            lon = city_data["lon"]
        else:
            print(f"Geocoding failed for {city}: {geo_json}")
            continue  # skip to next city

        # 2️Weather: get current weather
        weather_url = "https://api.openweathermap.org/data/2.5/weather"
        API_key_weather=os.getenv("API")
        weather_params = {"lat": lat, "lon": lon, "appid": API_key_weather, "units": "metric"}
        weather_response = requests.get(weather_url, params=weather_params)
        weather_json = weather_response.json()

        if weather_response.status_code == 200 and "main" in weather_json:
            temperature = weather_json["main"]["temp"]
            feels_like = weather_json["main"]["feels_like"]
            description = weather_json["weather"][0]["description"]
            humidity = weather_json["main"]["humidity"]
            wind_speed = weather_json["wind"]["speed"]

            # Append to data list
            data.append({
                "name": city,
                "country": country,
                "lat": lat,
                "lon": lon,
                "temp": temperature,
                "feels_like": feels_like,
                "weather": description,
                "humidity": humidity,
                "wind_speed": wind_speed
            })
        else:
            print(f"Weather fetch failed for {city}: {weather_json}")

    # 3️Build DataFrame
    df = pd.DataFrame(data)
    return df

# Example usage
cities = ["Berlin", "Hamburg", "Cologne"]
weather_df = get_weather_dataframe(cities)
print(weather_df)
```

```python
#We turn the above into a function
def push_weather_to_cloud(df, engine):
    # 1. Clean up the DataFrame to match SQL exactly
    df = df.rename(columns={
        "name": "city_name",
        "lat": "latitude",
        "lon": "longitude"
    })

    # 2. Add the timestamp (if not already there)
    from datetime import datetime
    df["weather_timestamp"] = datetime.now()

    # 3. Ensure city_id is handled
    # If you haven't mapped city_ids yet, we set it to None so MySQL can handle it 
    # or you can fill it if you have the IDs ready.
    if "city_id" not in df.columns:
        df["city_id"] = None

    # 5. Push to Cloud
    try:
        df.to_sql(
            "weather", 
            con=engine, 
            if_exists="append", # Use append to respect your AUTO_INCREMENT weather_id
            index=False
        )
        return("Cloud Update Successful!")
    except Exception as e:
        return(f"Export failed: {e}")
```

```python
import pandas as pd
import requests

# Example: list of airport IATA codes for your cities
iata_codes = ["BER", "CGN", "HAM"]  # Berlin, Cologne, Hamburg

# Store all arrivals in a single DataFrame
all_arrivals = []

for iata_code in iata_codes:
    url_arrivals = f"https://aerodatabox.p.rapidapi.com/flights/airports/iata/{iata_code}"
    querystring_arrivals = {
        "offsetMinutes": -120,
        "durationMinutes": 720,
        "withLeg": True,
        "direction": "Arrival",
        "withLocation": True
    }

    # Make the API request
    arrivals_response = requests.get(url_arrivals, headers=headers, params=querystring_arrivals)
    arrivals_data = arrivals_response.json()
    
    # Check if arrivals exist
    if "arrivals" in arrivals_data:
        arrivals_df = pd.json_normalize(
            arrivals_data['arrivals'],
            sep='_',
            meta=[
                ['flightNumber'],
                ['airline','name'],
                ['departure','airport','iata'],
                ['departure','airport','name'],
                ['departure','scheduledTimeLocal'],
                ['arrival','airport','icao'],
                ['arrival','airport','iata'],
                ['arrival','airport','name'],
                ['arrival','airport','location','lat'],
                ['arrival','airport','location','lon'],
                ['arrival','scheduledTimeLocal'],
                ['aircraft','model'],
                ['location','lat'],
                ['location','lon']
            ],
            errors='ignore'
        )

        # Rename columns for clarity
        arrivals_df = arrivals_df.rename(columns={
            'airline_name': 'airline_name',
            'departure_airport_iata': 'departure_airport_iata',
            'departure_airport_name': 'departure_airport_name',
            'departure_scheduledTimeLocal': 'departure_scheduledTimeLocal',
            'arrival_airport_icao': 'arrival_airport_icao',
            'arrival_airport_iata': 'arrival_airport_iata',
            'arrival_airport_name': 'arrival_airport_name',
            'arrival_airport_location_lat': 'arrival_airport_lat',
            'arrival_airport_location_lon': 'arrival_airport_lon',
            'arrival_scheduledTimeLocal': 'arrival_scheduledTimeLocal',
            'aircraft_model': 'aircraft_model',
            'location_lat': 'location_lat',
            'location_lon': 'location_lon'
        })

        # Optional: filter only useful columns
        useful_columns = [
            'flightNumber',
            'airline_name',
            'departure_airport_iata',
            'departure_airport_name',
            'departure_scheduledTimeLocal',
            'arrival_airport_icao',
            'arrival_airport_iata',
            'arrival_airport_name',
            'arrival_airport_lat',
            'arrival_airport_lon',
            'arrival_scheduledTimeLocal',
            'aircraft_model'
        ]
        arrivals_df = arrivals_df[[col for col in useful_columns if col in arrivals_df.columns]]

        # Append to main list
        all_arrivals.append(arrivals_df)

# Combine all arrivals into a single DataFrame
final_arrivals_df = pd.concat(all_arrivals, ignore_index=True)
print(final_arrivals_df)
```

```python
import requests
import pandas as pd
import os

# Your RapidAPI key
headers = {
    "X-RapidAPI-Key": os.getenv("FlightsAPI"),
    "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
}

def get_airports_by_city(lat, lon, radius_km=50):
    """
    Returns a DataFrame of airports near a city, with important info like ICAO/IATA and coordinates.
    """
    url = f"https://aerodatabox.p.rapidapi.com/airports/search/location/{lat}/{lon}/km/{radius_km}/16"
    querystring = {"withFlightInfoOnly": "true"}

    response = requests.get(url, headers=headers, params=querystring)
    response.raise_for_status()
    
    data = response.json()
    
    # Check if 'items' exists
    if "items" in data and data["items"]:
        # Normalize JSON into DataFrame
        airport_df = pd.json_normalize(
            data['items'],
            sep='.'
        )
        
        # Keep only desired columns
        useful_columns = [
            'icao',
            'iata',
            'name',
            'shortName',
            'municipalityName',
            'countryCode',
            'timeZone',
            'location.lat',
            'location.lon'
        ]
        airport_df = airport_df[[col for col in useful_columns if col in airport_df.columns]]
        return airport_df
    else:
        # Return empty DataFrame with columns if no airports found
        return pd.DataFrame(columns=[
            'icao',
            'iata',
            'name',
            'shortName',
            'municipalityName',
            'countryCode',
            'timeZone',
            'location.lat',
            'location.lon'
        ])

# Example: get airports near Berlin, Cologne, Munich
city_coords = {
    "Berlin": (52.5200, 13.4050),
    "Cologne": (50.9364, 6.9528),
    "Hamburg": (53.5500, 10.0000)
}

all_airports = pd.DataFrame()

for city, (lat, lon) in city_coords.items():
    df = get_airports_by_city(lat, lon)
    df['city'] = city  # add city column
    all_airports = pd.concat([all_airports, df], ignore_index=True)

print(all_airports)
```
